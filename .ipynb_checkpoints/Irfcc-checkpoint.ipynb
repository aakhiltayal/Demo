```python
def create_feff_amp_phase_file(rmin=0, rmax=10, data_group=None, path_feff=None, phase_name=None, R=3):
    xftr(data_group.r, 
         data_group.chir_mag, group=data_group, 
         qmax_out = 20, 
         rmin=rmin, rmax=rmax, window='sine', with_phase=True)
    
    path = "/Users/akhiltayal/Library/CloudStorage/GoogleDrive-akhil.tayal.bnl@gmail.com/My Drive/feff_amp_phases/"
    with open(path+'template_feff.dat', 'r') as template:
        temp = template.readlines()
        temp[0] = ' ' + phase_name + '                                          Feff8L (EXAFS)       0.1\n'
        temp[9] = f'   1   1.000   {R:1.4f}    2.6510  -18.44101 nleg, deg, reff, rnrmav(bohr), edge\n'
        
    hed = ''.join(temp)
    
    np.savetxt(path_feff+phase_name+"_custom.dat", np.column_stack((data_group.q,
                                                                   np.zeros(len(data_group.q)),
                                                                   data_group.chiq_mag,
                                                                   data_group.chiq_pha,
                                                                   np.zeros(len(data_group.q)),
                                                                   np.zeros(len(data_group.q)),
                                                                   np.zeros(len(data_group.q))  
                                                                  )),
               fmt = '%7.3f%12.4E%12.4E%12.4E%12.4E%12.4E%12.4E', header=hed, comments=''
              )
```
